# Graphical Representations for descriptive analysis

```python
# To find average rentals on weekends and weekdays

datafram_weekends=datafram[datafram['weekend']==1]
mean_ofweekends=datafram_weekends["rental"].mean()

datafram_weekdays=datafram[datafram['weekend']==0]
mean_ofweekdays=datafram_weekdays["rental"].mean()

meanrentals=[]
meanrentals.append(mean_ofweekends)
meanrentals.append(mean_ofweekdays)
x=["weekends(2016-2018)","weekdays(2016-2018)"]
import seaborn as sns
plt.title("Average rentals on weekends and weekdays")
ax=sns.barplot(y=meanrentals, x=x, palette="deep")
ax.set(ylabel=" Average rental")
plt.show()
```

```python
# Here we find the sum of rentals in each Year

year_rental=[]
year=[2016,2017,2018]
year_rental.append(datafram[datafram['year']==2016]['rental'].sum())
year_rental.append(datafram[datafram['year']==2017]['rental'].sum())
year_rental.append(datafram[datafram['year']==2018]['rental'].sum())
axes=sns.barplot(y=year_rental, x=year, palette="deep")
axes.set(xlabel="Year",ylabel="rental")
plt.title("Total rental variation over year 2016,2017,2018")
plt.show()
```

```python
# To find the average rentals over the year and to depict which year had maximum rental

year_rental_avg=[]
year_rental_avg.append(datafram[datafram['year']==2016]['rental'].mean())
year_rental_avg.append(datafram[datafram['year']==2017]['rental'].mean())
year_rental_avg.append(datafram[datafram['year']==2018]['rental'].mean())
year_rental_avg
```

```python

Axis=sns.barplot(y=year_rental_avg, x=year, palette="deep")
Axis.set(ylabel="Average Rental", xlabel="Year")
plt.title("Average rental over each year")
```

```python
# Here we can see the variation of rental with time during weekdays and weekend
sns.lineplot(data=datafram, x="hour", y="rental",hue='weekend')
plt.title("Rentals variation during weekend")
```

```python
#Fromm this plot we can see whether the temperature plays an important role in variation of rental

sns.histplot(data=datafram, x="temperature", y="rental",hue='weekend')
plt.title("Temperature variation on weekdays and weekends")
```

```python
# Here we can Check whether the humidty plays an important part of rental variation
sns.lineplot(data=datafram, x="humidity", y="rental")
```

```python
# from this plot we can see relation between Rental variations as weather changes
sns.lineplot(data=datafram, x="hour", y="rental",hue='weather')
plt.title("Affect of weather on rental")
```

```python
# Windspeed affect on rental

sns.scatterplot(data=datafram, x="windspeed", y="rental",hue='hour')
plt.title("Affect of windspeed on rental")
```

```python
#Here using boxplot and catplot we can see how atmospheric conditions vary across the data

plt.rcParams["figure.figsize"] = (10,8)
sns.boxplot(data= datafram.loc[:,['temperature','dewpoint','humidity','windspeed','windgust','pressure','precip','precipaccum']])
sns.catplot(height=8,data= datafram.loc[:,['temperature','dewpoint','humidity','windspeed','windgust','pressure','precip','precipaccum'                                 ]])
plt.show()
```

```python
#Rental variation over season which helps us to determine in which season there is maximum use
sns.lineplot(data=datafram, x="hour", y="rental",hue='season')
plt.title("Rental variation over season")
```

```python
# Here we can see the distribution of probability density function of rentals

sns.displot(data=datafram, x="rental", kind="kde")
plt.title("probability density function of rentals")
plt.show()
```

```python
#Monthly rental variation over the years, we can check how there is variation of rentals over the years

month_rental=[]
month=[x+1 for x in range(12)]
for i in range(12):
    month_rental.append(datafram[datafram['month']==i+1]['rental'].sum())

sns.barplot(y=month_rental, x=month, palette="deep")
plt.title("Total month rental variation over year 2016,2017,2018")
```

```python
# We can check the hourly rental variation over the years

hour_rental=[]
hour=[x for x in range(24)]
for i in range(24):
    hour_rental.append(datafram[datafram['hour']==i]['rental'].sum())
    
sns.barplot(y=hour_rental, x=hour, palette="deep")
plt.title("Total hourly rental variation over year 2016,2017,2018")
```

# Model 1 Creation Using Linear Regression

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
#converstion of characters in season to binary values
from sklearn.preprocessing import OneHotEncoder
onehotencoder = OneHotEncoder()
xx = onehotencoder.fit_transform(np.array(datafram['season']).reshape((-1,1))).toarray()
```

```python
xx
```

```python
datafram[['fall','winter','spring','summer']]=xx
```

```python
datafram
```

```python
# Now we split data for Training and Testing using train_test_split
# Also for that we drop rental and season 
from sklearn.model_selection import train_test_split
Y=datafram['rental']
X=datafram.drop(['rental','season'],axis=1)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 1,shuffle=False)
```

```python
# Training the model by excluding rental and season
model=LinearRegression()
model.fit(X_train,Y_train)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
# Calculating MSE for trained model data
y_predi=model.predict(X_train)
mean_squared_error(y_predi,Y_train)
```

```python
# Calculating MAE for trained data
mean_absolute_error(y_predi,Y_train)
```

```python

y_pred=model.predict(X_test)
```

```python
# Calculating MSE for test model data
mean_squared_error(y_pred,Y_test)
```

```python
# Calculating MAE for test data
mean_absolute_error(y_pred,Y_test)
```

```python
model.coef_
```

```python
# we are finding the least important feature which affects rentals
np.argmin(model.coef_)
```

```python
# we are finding the maximum important feature which affects rentals
np.argmax(model.coef_)
```

```python

model.score(X_test,Y_test)
```
